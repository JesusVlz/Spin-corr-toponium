## Funciones auxiliares

````python
def add_text_labels(pad, title="", info="", lumi=""):
    """Añadir etiquetas de texto en el pad"""
    pad.cd()
    
    # Título superior izquierdo
    if title:
        latex_title = ROOT.TLatex()
        latex_title.SetNDC()
        latex_title.SetTextFont(42)
        latex_title.SetTextSize(0.05)
        latex_title.DrawLatex(0.13, 0.92, title)
    
    # Info superior derecho
    if info:
        latex_info = ROOT.TLatex()
        latex_info.SetNDC()
        latex_info.SetTextFont(42)
        latex_info.SetTextSize(0.04)
        latex_info.SetTextAlign(31)  # Alineado a la derecha
        latex_info.DrawLatex(0.95, 0.95, info)
    
    # Luminosidad (debajo de info)
    if lumi:
        latex_lumi = ROOT.TLatex()
        latex_lumi.SetNDC()
        latex_lumi.SetTextFont(42)
        latex_lumi.SetTextSize(0.035)
        latex_lumi.SetTextAlign(31)
        latex_lumi.DrawLatex(0.95, 0.87, lumi)

def create_legend_with_stats(h_data, h1, h2, label_data, label1, label2, 
                             x1=0.15, y1=0.70, x2=0.38, y2=0.90):
    """Crear leyenda con estadísticas solo para MC"""
    legend = ROOT.TLegend(x1, y1, x2, y2)
    legend.SetBorderSize(0)
    legend.SetFillStyle(0)
    legend.SetTextFont(42)
    legend.SetTextSize(0.035)
    
    # Datos experimentales - solo etiqueta, sin estadísticas
    legend.AddEntry(h_data, label_data, "ep")
    
    # Primera simulación con estadísticas
    mean1 = h1.GetMean()
    sigma1 = h1.GetStdDev()
    legend.AddEntry(h1, label1, "l")
    legend.AddEntry(ROOT.nullptr, f"#mu = {mean1:.2f}, #sigma = {sigma1:.2f}", "")
    
    # Segunda simulación con estadísticas
    mean2 = h2.GetMean()
    sigma2 = h2.GetStdDev()
    legend.AddEntry(h2, label2, "l")
    legend.AddEntry(ROOT.nullptr, f"#mu = {mean2:.2f}, #sigma = {sigma2:.2f}", "")
    
    return legend

def create_ratio_graph_from_mc_and_data(h_mc, h_data_graph, name):
    """Crear TGraphErrors de ratio Data/MC con propagación de errores correcta"""
    import math
    
    ratio_graph = ROOT.TGraphErrors()
    n_points = h_data_graph.GetN()
    
    point_idx = 0
    for i in range(n_points):
        # Obtener punto del data (TGraphAsymmErrors)
        x_data = h_data_graph.GetPointX(i)
        y_data = h_data_graph.GetPointY(i)
        
        # Error del data (promedio de asimétricos)
        error_data_high = h_data_graph.GetErrorYhigh(i)
        error_data_low = h_data_graph.GetErrorYlow(i)
        error_data = (error_data_high + error_data_low) / 2.0
        
        # Encontrar el bin correspondiente en el histograma MC
        bin_mc = h_mc.FindBin(x_data)
        y_mc = h_mc.GetBinContent(bin_mc)
        error_mc = h_mc.GetBinError(bin_mc)
        
        # Calcular ratio Data/MC
        if y_mc != 0 and y_data > 0:
            ratio = y_data / y_mc
            
            # Propagación de errores: error_ratio = ratio * sqrt((error_data/y_data)^2 + (error_mc/y_mc)^2)
            error_ratio = ratio * math.sqrt(pow(error_data / y_data, 2) + pow(error_mc / y_mc, 2))

            print(f"Point {i}: x={x_data:.3f}, Data={y_data:.3f}±{error_data:.3f}, MC={y_mc:.3f}±{error_mc:.3f}, Ratio={ratio:.3f}±{error_ratio:.3f}")
            
            ratio_graph.SetPoint(point_idx, x_data, ratio)
            ratio_graph.SetPointError(point_idx, 0, error_ratio)
            point_idx += 1
        else:
            print(f"Warning: Division by zero at point {i}, x={x_data:.3f}")
    
    return ratio_graph
````

## Iterar sobre todos los histogramas

````python
# Contadores
processed_count = 0
failed_histograms = []

# Iterar sobre cada configuración de histograma
for idx, config in enumerate(histogram_configs, 1):
    # Desempaquetar configuración (con título X opcional)
    if len(config) == 5:
        mc_hist_name, hepdata_file, hepdata_hist_name, custom_x_title, custom_y_title = config
    else:
        mc_hist_name, hepdata_file, hepdata_hist_name = config
        custom_x_title = None  # Usar título original del histograma
        custom_y_title = None  # Usar título original del histograma
    
    print(f"\n{'='*60}")
    print(f"Procesando [{idx}/{len(histogram_configs)}]: {mc_hist_name}")
    print(f"  Archivo HEPData: {hepdata_file}")
    print(f"  Histograma en HEPData: {hepdata_hist_name}")
    if custom_x_title:
        print(f"  Título eje X personalizado: {custom_x_title}")
    if custom_y_title:
        print(f"  Título eje Y personalizado: {custom_y_title}")
    print(f"{'='*60}")
    
    try:
        # Obtener los histogramas de las simulaciones
        h1 = file1.Get(mc_hist_name)
        h2 = file2.Get(mc_hist_name)
        
        # Verificar existencia en simulaciones
        if not h1:
            raise Exception(f"No se encontró '{mc_hist_name}' en {file1_path}")
        if not h2:
            raise Exception(f"No se encontró '{mc_hist_name}' en {file2_path}")
        
        # Abrir archivo de HEPData específico para este histograma
        file_data = ROOT.TFile.Open(hepdata_file)
        if not file_data or file_data.IsZombie():
            raise Exception(f"No se pudo abrir el archivo HEPData: {hepdata_file}")
        
        # Obtener histograma de HEPData
        h_data = file_data.Get(hepdata_hist_name)
        if not h_data:
            raise Exception(f"No se encontró '{hepdata_hist_name}' en {hepdata_file}")
        
        # Clonar histogramas
        h1 = h1.Clone(f"h1_clone_{idx}")
        h2 = h2.Clone(f"h2_clone_{idx}")
        h_data = h_data.Clone(f"h_data_clone_{idx}")
        h1.SetDirectory(0)
        h2.SetDirectory(0)
        # h_data.SetDirectory(0)  # TGraph no tiene SetDirectory
        
        # Cerrar archivo de HEPData después de clonar
        file_data.Close()
        
        print(f"  Simulación 1: {h1.GetEntries():.0f} entradas")
        print(f"  Simulación 2: {h2.GetEntries():.0f} entradas")
        print(f"  Datos CMS: {h_data.GetN()} puntos")
        
        # ==========================================
        # CONFIGURAR ESTILOS
        # ==========================================
        
        # Simulación 1 (azul)
        h1.SetLineColor(ROOT.kBlue)
        h1.SetLineWidth(2)
        h1.SetLineStyle(1)
        
        # Simulación 2 (rojo)
        h2.SetLineColor(ROOT.kRed)
        h2.SetLineWidth(2)
        h2.SetLineStyle(1)
        
        # Datos experimentales (negro, con puntos)
        h_data.SetLineColor(ROOT.kBlack)
        h_data.SetLineWidth(2)
        h_data.SetMarkerColor(ROOT.kBlack)
        h_data.SetMarkerStyle(20)
        h_data.SetMarkerSize(1.0)
        
        # ==========================================
        # CANVAS CON RATIO PLOT
        # ==========================================
        canvas = ROOT.TCanvas(f"canvas_{idx}", f"Comparación: {mc_hist_name}", 800, 800)
        
        # Pad superior (histogramas)
        pad1 = ROOT.TPad("pad1", "pad1", 0, 0.3, 1, 1.0)
        pad1.SetBottomMargin(0.02)
        pad1.SetTopMargin(0.08)
        pad1.SetLeftMargin(0.13)
        pad1.SetRightMargin(0.05)
        pad1.Draw()
        pad1.cd()
        
        # Determinar rango Y
        # Para TGraphAsymmErrors, necesitamos calcular el máximo manualmente
        if h_data.InheritsFrom("TGraph"):
            # Obtener el máximo del TGraphAsymmErrors considerando errores
            max_data = 0
            for i in range(h_data.GetN()):
                y_val = h_data.GetPointY(i)
                y_err_high = h_data.GetErrorYhigh(i)
                max_data = max(max_data, y_val + y_err_high)
            max_val = max(h1.GetMaximum(), h2.GetMaximum(), max_data)
        else:
            max_val = max(h1.GetMaximum(), h2.GetMaximum(), h_data.GetMaximum())
        
        h_data.SetMaximum(1.8)
        h_data.SetMinimum(0.0)

        if custom_y_title:
            y_title = custom_y_title
        else:
            y_title = h1.GetYaxis().GetTitle()       

        # Dibujar histogramas (datos primero)
        h_data.GetYaxis().SetTitle(y_title)
        h_data.GetYaxis().SetTitleSize(0.05)
        h_data.GetYaxis().SetTitleOffset(1.3)
        h_data.GetYaxis().SetLabelSize(0.04)
        
        # Ocultar etiquetas del eje X en el pad superior
        h_data.GetXaxis().SetLabelSize(0)
        h_data.GetXaxis().SetTitleSize(0)
        
        h_data.Draw("AP")  # Datos con barras de error (A = ejes, P = puntos)
        h1.Draw("HIST SAME")  # Simulación 1 (histograma)
        h2.Draw("HIST SAME")  # Simulación 2 (histograma)
        h_data.Draw("P SAME")  # Redibujar datos encima
        
        # Leyenda con estadísticas
        legend = create_legend_with_stats(h_data, h1, h2, label_data, label1, label2)
        legend.Draw()
        
        # Añadir etiquetas de texto
        add_text_labels(pad1, plot_title, collider_info, luminosity_info)
        
        # ==========================================
        # PAD INFERIOR (RATIO)
        # ==========================================
        canvas.cd()
        
        pad2 = ROOT.TPad("pad2", "pad2", 0, 0.0, 1, 0.3)
        pad2.SetTopMargin(0.02)
        pad2.SetBottomMargin(0.30)
        pad2.SetLeftMargin(0.13)
        pad2.SetRightMargin(0.05)
        pad2.Draw()
        pad2.cd()
        
        # Calcular ratios usando TGraphErrors (Data / MC)
        ratio_graph1 = create_ratio_graph_from_mc_and_data(h1, h_data, f"ratio1_{idx}")
        ratio_graph2 = create_ratio_graph_from_mc_and_data(h2, h_data, f"ratio2_{idx}")
        
        # Configurar estilos de los ratios
        ratio_graph1.SetLineColor(ROOT.kBlue)
        ratio_graph1.SetMarkerColor(ROOT.kBlue)
        ratio_graph1.SetMarkerStyle(20)
        ratio_graph1.SetMarkerSize(0.8)
        
        ratio_graph2.SetLineColor(ROOT.kRed)
        ratio_graph2.SetMarkerColor(ROOT.kRed)
        ratio_graph2.SetMarkerStyle(21)
        ratio_graph2.SetMarkerSize(0.8)
        
        # Dibujar primer ratio con ejes
        ratio_graph1.Draw("APE")
        
        # Configurar ejes del ratio
        ratio_graph1.GetYaxis().SetTitle("Data/MC")
        ratio_graph1.GetYaxis().SetTitleSize(0.12)
        ratio_graph1.GetYaxis().SetTitleOffset(0.5)
        ratio_graph1.GetYaxis().SetLabelSize(0.08)
        ratio_graph1.GetYaxis().SetNdivisions(505)
        
        ratio_graph1.GetXaxis().SetTitleSize(0.12)
        ratio_graph1.GetXaxis().SetTitleOffset(1.0)
        ratio_graph1.GetXaxis().SetLabelSize(0.10)
        
        # Usar título personalizado si se proporcionó, sino usar el original
        if custom_x_title:
            x_title = custom_x_title
        else:
            x_title = h1.GetXaxis().GetTitle()
        ratio_graph1.GetXaxis().SetTitle(x_title)
        
        # Rango del ratio
        ratio_graph1.SetMinimum(0.40)
        ratio_graph1.SetMaximum(4.0)
        
        # Dibujar segundo ratio encima
        ratio_graph2.Draw("PE SAME")
        
        # Línea horizontal en 1
        x_min = h1.GetXaxis().GetXmin()
        x_max = h1.GetXaxis().GetXmax()
        line = ROOT.TLine(x_min, 1, x_max, 1)
        line.SetLineColor(ROOT.kGray+1)
        line.SetLineStyle(2)
        line.SetLineWidth(2)
        line.Draw()
        
        # Banda de incertidumbre (opcional, a 10%)
        band_up = ROOT.TLine(x_min, 1.1, x_max, 1.1)
        band_up.SetLineColor(ROOT.kGray)
        band_up.SetLineStyle(3)
        band_up.Draw()
        
        band_down = ROOT.TLine(x_min, 0.9, x_max, 0.9)
        band_down.SetLineColor(ROOT.kGray)
        band_down.SetLineStyle(3)
        band_down.Draw()
        
        canvas.Update()
        
        # Guardar
        output_file = os.path.join(output_folder, f"{mc_hist_name}_data_comparison.pdf")
        canvas.SaveAs(output_file)
        print(f"  ✓ Guardado: {output_file}")
        
        processed_count += 1
        
        # Limpiar memoria
        del canvas, pad1, pad2, h1, h2, h_data, ratio_graph1, ratio_graph2, legend
        
    except Exception as e:
        print(f"  ✗ ERROR: {str(e)}")
        failed_histograms.append((mc_hist_name, str(e)))
        continue

# Resumen final
print(f"\n{'='*60}")
print(f"RESUMEN")
print(f"{'='*60}")
print(f"Procesados exitosamente: {processed_count}/{len(histogram_configs)}")
print(f"Archivos guardados en: {output_folder}/")

if failed_histograms:
    print(f"\n⚠ Histogramas fallidos: {len(failed_histograms)}")
    for name, error in failed_histograms:
        print(f"  - {name}: {error}")
````

## Notas sobre el estilo

Este notebook genera plots comparando simulaciones con datos experimentales:

### Panel superior:
- **Datos CMS**: Puntos negros con barras de error
- **Simulación 1**: Línea azul (histograma)
- **Simulación 2**: Línea roja (histograma)
- **Leyenda**: Incluye μ y σ para cada distribución
- **Etiquetas**: Título, √s, y luminosidad

### Panel inferior (Ratio):
- **MC/Data**: Ratio de simulaciones respecto a datos
- **Línea de referencia**: y=1 (línea gris punteada)
- **Bandas**: ±10% de incertidumbre (líneas grises)
- **Azul**: Simulación 1 / Datos
- **Rojo**: Simulación 2 / Datos
